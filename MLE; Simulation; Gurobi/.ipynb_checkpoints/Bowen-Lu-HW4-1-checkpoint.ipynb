### Problem 1. 
Tab "Solenoid Valve Data" has observations of time-to-failure in years of solenoid valves used in this kind of reactor application. We are concerned in an 8-year lifetime of the reactor process. Use the Python to solve the following problems:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_csv("Solenoid Valve Data.csv", header=0)
data = list(data.iloc[:,0])
```

#### P1.1: Find a best Weibull distribution fit to the data.

```python
from scipy.stats import weibull_min

# find the best Weibull distribution fit to the data
shape1, loc1, scale1 = weibull_min.fit(data)
print(shape1, loc1, scale1)
```

```python
plt.hist(data, normed=True)
    
xt = plt.xticks()[0]  
xmin, xmax = min(xt), max(xt)  
lnspc = np.linspace(xmin, xmax, len(data))
    
pdf = weibull_min.pdf(lnspc, shape1, loc1, scale1)
plt.plot(lnspc, pdf, label="Norm")
plt.show()
```

#### P1.2: Find a best Exponential distribution fit to the data.

```python
from scipy.stats import expon

# find the best Exponential distribution fit to the data
loc2, scale2 = expon.fit(data)
print(loc2, scale2)
```

```python
plt.hist(data, normed=True)
    
xt = plt.xticks()[0]  
xmin, xmax = min(xt), max(xt)  
lnspc = np.linspace(xmin, xmax, len(data))
    
pdf = expon.pdf(lnspc, loc2, scale2)
plt.plot(lnspc, pdf, label="Norm")
plt.show()
```

#### P1.3: Find a best Lognormal distribution fit to the data.

```python
from scipy.stats import lognorm

# find the best Lognormal distribution fit to the data
shape3, loc3, scale3 = lognorm.fit(data)
print(shape3, loc3, scale3)
```

```python
plt.hist(data, normed=True)
    
xt = plt.xticks()[0]  
xmin, xmax = min(xt), max(xt)  
lnspc = np.linspace(xmin, xmax, len(data))
    
pdf = lognorm.pdf(lnspc, shape3, loc3, scale3)
plt.plot(lnspc, pdf, label="Norm")
plt.show()
```

#### (I find that fitting data without arguments like loc=0 or floc=0 can obtain the best model with minimum Chi-square or AIC or BIC, which could get different parameter estimates compared with common solutions)

### Problem 2.
Use Chi-Square Goodness of Fit, AIC, and BIC to select a best model for our data out of the three obtained in Problem 1.

```python
# Calculate Chi-square Goodness of Fit for each model
from scipy.stats import chisquare

hist, bin_edges = np.histogram(data, bins='auto')
observed_values = hist

cdf1 = weibull_min.cdf(bin_edges, shape1, loc1, scale1)
expected_values1 = len(data)*np.diff(cdf1)
print("Chi-Square of Weibull:", chisquare(observed_values, expected_values1)[0], "(p-value=", chisquare(observed_values, expected_values1)[1])

cdf2 = expon.cdf(bin_edges, loc2, scale2)
expected_values2 = len(data)*np.diff(cdf2)
print("Chi-Square of Exponential:", chisquare(observed_values, expected_values2)[0], "(p-value=", chisquare(observed_values, expected_values2)[1])

cdf3 = lognorm.cdf(bin_edges, shape3, loc3, scale3)
expected_values3 = len(data)*np.diff(cdf3)
print("Chi-Square of Lognormal:", chisquare(observed_values, expected_values3)[0], "(p-value=", chisquare(observed_values, expected_values3)[1])
```

```python
# Calculate AIC for each model

lnL1 = 0
lnL2 = 0
lnL3 = 0

for i in data:
    lklh1 = np.log(weibull_min.pdf(i, shape1, loc1, scale1))
    lnL1 += lklh1
    lklh2 = np.log(expon.pdf(i, loc2, scale2))
    lnL2 += lklh2
    lklh3 = np.log(lognorm.pdf(i, shape3, loc3, scale3))
    lnL3 += lklh3

AIC1 = 2*3-2*lnL1
AIC2 = 2*2-2*lnL2
AIC3 = 2*3-2*lnL3

print("AIC of Weibull:", AIC1)
print("AIC of Exponential:", AIC2)
print("AIC of Lognormal:", AIC3)
```

```python
# Calculate BIC for each model

BIC1 = np.log(len(data))*3-2*lnL1
BIC2 = np.log(len(data))*2-2*lnL2
BIC3 = np.log(len(data))*3-2*lnL3

print("BIC of Weibull:", BIC1)
print("BIC of Exponential:", BIC2)
print("BIC of Lognormal:", BIC3)
```

#### Conclusion: Model of Weibull distribution achieves the minimum Chi-Square Goodness of Fit, AIC and BIC, so it is selected as the best model.

### Problem 8.
Use Gurobi and Python to pose and solve Problem 6.

```python
from gurobipy import *

# Create a new model and its variables
m = Model("ReachAcceptability")

x1 = m.addVar(vtype=GRB.BINARY, name="x1")
x2 = m.addVar(vtype=GRB.BINARY, name="x2")
x3 = m.addVar(vtype=GRB.BINARY, name="x3")
x4 = m.addVar(vtype=GRB.BINARY, name="x4")

m.update
```

```python
# Set the objective
m.setObjective(80*x1 + 10*x2 + 20*x3 + 70*x4 , GRB.MINIMIZE)

# Add constraints
m.addConstr((0.13*(1-(1-1/1.18)*x1)+0.08*(1-(1-1/2)*x2)) * (0.17*(1-(1-1/1.10)*x3)+0.127*(1-(1-1/1.25)*x4)) <= 0.03, "c0")

m.update
```

```python
# Optimize model
m.optimize()
```

```python
# Print optimal value
for v in m.getVars():
    print('%s %g' % (v.varName, v.x))
    
print('Obj: %g' % m.objVal)
```
