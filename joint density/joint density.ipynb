## Problem 1
__(a)__ In order for f(x,y) to be a legitimate joint density function, the following must be true: 
<br>
<br>
$$ \int_{-\infty}^{\infty} \int_{-\infty}^{\infty} c e^{-x-y} dx dy= 1 $$
<br>
<br>
Therefore, the value of c that makes this a legitimate joint density function will be that which makes the cumulative sum of f(x,y) evaluated at each x and y equal to 1. This is derived below:
<br>
$$ \int_{0}^{\infty} \int_{0}^{\infty} c e^{-x-y} dx dy= 1 $$
<br>
$$ c \int_{0}^{\infty} e^{-x}dx \int_{0}^{\infty} e^{-y}dy= 1 $$
<br>
$$ c \left[\frac{e^{-x}}{-1}\right]_{0}^{\infty} \left[\frac{e^{-y}}{-1}\right]_{0}^{\infty} = 1$$
<br>
$$ c \left[\frac{0-1}{-1}\right] \left[\frac{0-1}{-1}\right] = 1 $$
<br>
$$ c = 1 $$

__(b)__ The probability of the first risk exceeding the second is:
$$ P(X>Y) = \int_{0}^{\infty} \int_{y}^{\infty} c e^{-x-y} dx dy $$
<br>
$$ = \int_{0}^{\infty} e^{-y} \left[\int_{y}^{\infty} e^{-x} dx \right] dy $$
<br>
$$ = \int_{0}^{\infty} e^{-y} \left[\left[\frac{e^{-x}}{-1} \right]_{y}^{\infty}\right] dy $$
<br>
$$ = \int_{0}^{\infty} e^{-y} \left[\frac{0-e^{-y}}{-1} \right] dy $$
<br>
$$ = \int_{0}^{\infty} e^{-2y} dy $$
<br>
$$ = \left[\frac{e^{-2y}}{-2} \right]_{0}^{\infty} $$
<br>
$$ = \frac{0-1}{-2} = 0.5 $$
<br>
The probability of the first risk exceeding the second is 50%.

__(c)__ $$ P(X+Y>1) = 1 - P(X+Y<1) $$
<br>
$$ P(X+Y<1) = \int_{0}^{1} \int_{0}^{1-y} c e^{-x-y} dx dy $$
<br>
$$ = \int_{0}^{1} e^{-y} \left[\left[\frac{e^{-x}}{-1}\right]_{0}^{1-y} \right] dy$$
<br>
$$ = \int_{0}^{1} e^{-y} \left[\frac{e^{-(1-y)}-1}{-1} \right] dy$$
<br>
$$ = \int_{0}^{1} \left[e^{-y} - e^{-y-1+y} \right] dy$$
<br>
$$ = \int_{0}^{1} e^{-y}dy - e^{-1}\int_{0}^{1}dy $$
<br>
$$ = \left[\frac{e^{-y}}{-1} \right]_{0}^{1} - e^{-1}\left[y\right]_{0}^{1} $$
<br>
$$ = \frac{e^{-1}-1}{-1} - e^{-1} $$
<br>
$$ = 1 - 2e^{-1} $$
<br>
$$ \therefore P(X+Y>1) = 1 - \left[1-\frac{2}{e} \right] = \frac{2}{e} = 0.7357 $$
<br>
The probability of the sum of the two risks exceeding 1 is 73.57%.
